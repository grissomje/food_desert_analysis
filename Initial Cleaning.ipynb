```python
import pandas as pd
from functools import reduce
```

```python
county= pd.read_excel('FoodEnvironmentAtlas.xls', sheet_name='Supplemental Data - County')
access=pd.read_excel('FoodEnvironmentAtlas.xls', sheet_name='ACCESS')
stores=pd.read_excel('FoodEnvironmentAtlas.xls', sheet_name='STORES')
restaurants=pd.read_excel('FoodEnvironmentAtlas.xls', sheet_name='RESTAURANTS')
insecurity=pd.read_excel('FoodEnvironmentAtlas.xls', sheet_name='INSECURITY')
local=pd.read_excel('FoodEnvironmentAtlas.xls', sheet_name='LOCAL')
socioeconomic=pd.read_excel('FoodEnvironmentAtlas.xls', sheet_name='SOCIOECONOMIC')
```

```python
#county columns to drop
county_drop=['Population_Estimate_2016','Population_Estimate_2017','Population_Estimate_2018']

county = county.drop(columns=county_drop)

#access columns to drop
access_drop=['LACCESS_CHILD10','LACCESS_CHILD15','LACCESS_CHILD_10_15','PCT_LACCESS_CHILD10','PCT_LACCESS_CHILD15','LACCESS_SENIORS10','LACCESS_SENIORS15','PCH_LACCESS_SENIORS_10_15','PCT_LACCESS_SENIORS10','PCT_LACCESS_SENIORS15','LACCESS_WHITE15','PCT_LACCESS_WHITE15','LACCESS_BLACK15','PCT_LACCESS_BLACK15','LACCESS_HISP15','PCT_LACCESS_HISP15','LACCESS_NHASIAN15','PCT_LACCESS_NHASIAN15','LACCESS_NHNA15','PCT_LACCESS_NHNA15','LACCESS_NHPI15','PCT_LACCESS_NHPI15','LACCESS_MULTIR15','PCT_LACCESS_MULTIR15']

access = access.drop(columns=access_drop)

#store columns to drop
stores_drop=['SNAPS12','SNAPS17','PCH_SNAPS_12_17','SNAPSPTH12','SNAPSPTH17','PCH_SNAPSPTH_12_17','WICS11','WICS16','PCH_WICS_11_16','WICSPTH11','WICSPTH16','PCH_WICSPTH_11_16']

stores=stores.drop(columns=stores_drop)
```

```python
dfs=[county,access,stores,restaurants,insecurity,local,socioeconomic]

df_final = reduce(lambda left,right: pd.merge(left,right,on=['FIPS']), dfs)
df_final
```

```python
remove=['_10_15','16','17','18']
for r in remove:
    df = df_final[df_final.columns.drop(list(df_final.filter(regex=r)))]

df
```

```python
df = df.loc[:,~df.columns.duplicated()]
```

```python
df = df.drop(columns=['State_y','County_y','County','State'])
```

```python
df.to_csv('food_access.csv')
```
